```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline  
import os
```

```python
path = './drive/MyDrive/ColabNotebooks/AppliedDeepLearning/hyperparameter_logs/'
```

```python
for i in range(1,19):
  if i != 10:

    train = []
    valid = []

    with open(path + 'segmentation_notebook_experiment_' + str(i) + '/metrics.txt') as f:
      [train.append(float(line[line.rfind('=') + 1:])) for line in f.readlines() if 'train' in line]
    with open(path + 'segmentation_notebook_experiment_' + str(i) + '/metrics.txt') as f:
      [valid.append(float(line[line.rfind('=') + 1:])) for line in f.readlines() if 'valid' in line]

    losses = pd.DataFrame(
        {'epoch': range(len(train)),
        'train': train,
        'valid': valid
        })
    
    losses.plot(x="epoch", y=["train", "valid"], ylabel='BCE Dice Loss', title='Experiment ' + str(i))
    plt.savefig(path + 'segmentation_notebook_experiment_' + str(i) + '/loss_plot.jpg')
```
